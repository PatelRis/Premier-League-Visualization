<p style="font-size:18px; font-family:verdana; line-height: 1.7em">The Premier League, often referred to as the English Premier League or the EPL, is the top level of the English football league system. Contested by 20 clubs, it operates on a system of promotion and relegation with the English Football League. Seasons run from August to May with each team playing 38 matches.</p>

 <p style="font-size:25px; font-family:verdana; line-height: 1.7em"> 
    <b>Attributes In The Dataset</b></p>

<p style="font-size:18px; color:#ce9965; font-family:verdana;">
<code>Season</code>: Football season year <br>
<code>Pos</code>: Final position of the team for that season<br>
<code>Team</code>: Name of the team<br>
<code>Pld</code>: Number of games played<br>
<code>W</code>: Win<br>
<code>D</code>: Draw<br>
<code>L</code>: Loss<br>
<code>GF</code>: Goals Scored<br>
<code>GA</code>: Goals Conceded<br>
<code>GD</code>: Goal Difference<br>
<code>Pts</code>: Total Points scored by the team in that season<br>
<code>Qualification or relegation</code>: Whether the team has qualified for the european continental cup or has been relegated to a lower tier league

```python
import pandas as pd # data analysis
import numpy as np # linear algebra

#import libraries for data visualization
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots

import warnings
warnings.filterwarnings('ignore')

data = pd.read_csv('../input/premier-league-standings-11-seasons-20102021/EPL Standings 2010-2021.csv')
data
```

```python
fig = px.sunburst(data, path=['Team','Pld', 'W' , 'D', 'L'])
fig.update_layout (height = 1600 , width = 1600)
fig.show()
```

 <p style="font-size:35px; font-family:verdana; line-height: 1.7em"> 
    <b>Most Number Of Wins In Premier League</b></p>

```python
Most_wins = data.loc[(data['Pos'] == 1)]
Most_wins.head(10)
```

```python
Most_number_of_wins = Most_wins['Team'].value_counts()
Most_number_of_wins
```

 <p style="font-size:35px; font-family:verdana; line-height: 1.7em"> 
    <b>Season: 2010-11</b></p>

```python
Data_From_2010_To_2011_loc = data.loc[(data['Season'] == '2010-11')]
Data_From_2010_To_2011_loc
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(
    x=Data_From_2010_To_2011_loc['Team'],
    y=Data_From_2010_To_2011_loc['Pld'],
    name='Matches played',
    marker_color='mediumvioletred'
))
fig.add_trace(go.Bar(
    x=Data_From_2010_To_2011_loc['Team'],
    y=Data_From_2010_To_2011_loc['W'],
    name='Matches win',
    marker_color='springgreen'
))
fig.add_trace(go.Bar(
    x=Data_From_2010_To_2011_loc['Team'],
    y=Data_From_2010_To_2011_loc['D'],
    name='Matches draw',
    marker_color='dodgerblue'
))
fig.add_trace(go.Bar(
    x=Data_From_2010_To_2011_loc['Team'],
    y=Data_From_2010_To_2011_loc['L'],
    name='Matches loss',
    marker_color='darkblue'
))

fig.update_layout(title="Matches Played-Win-Draw-Loss",
                  barmode='stack',
                  paper_bgcolor= '#FFFDE7',
                  plot_bgcolor= '#FFFDE7',
                  title_font_size=30,
                  xaxis_tickangle=-45)

fig.show()
```
